Control Flow with Conditionals.

Often times, we'll want our programs to execute some code based on a condition.

Some psuedo-code to explain:

```
if some_condition is True:
    do_stuff()
else:
    do_nothing()
```

````python
# let's make a variable to see if-else statements in action
score = 72
````

````python
72 >= 70
````

````python
# if score is greater than or equal to some value, print Pass!, otherwise print Fail
if score >= 70:
    print("Yay you passed!")
else:
    print("Oh no, you did not pass... better luck next time")
````

What happens if there are multiple conditions to check?

We can use `elif` to check conditions *after* the `if` statement.

````python
# First, what happens if we do a bunch of if statements?
if score >= 90:
    print("You got an A!")
if score >= 80:
    print("You got a B!")
if score >= 70:
    print("You got a C!")
if score >= 60:
    print("You got a D :(")
else:
    print("Sorry, you failed")
````

So why did it print that we got both a C and a D?

Because both of those conditions evaluate to `True`!

We typically only want one condition to be met in our `if-elif-else` statements. 

Let's fix the above to reflect that.

We also have `and` and `or`

`and` can be used to check multiple conditions that are `True`. 

`and` is very strict logic

## Loops

Often times, we will want to repeat a task a certain amount of times or until some condidtion is not met. 

There are two primary loops you will see in python.

`for` loops are when you want to iterate over an iterable.

Okay, so what does that mean?

Here is some psuedo-code to showcase:

```
for thing in things:
    
    do_stuff(thing)
```
Let's showcase for loops with your `name` variable defined earlier and then also introduce the built-in `range()` function.

````python
# re-defining
name = 'Hank'
````

````python
for i in name:
    print(i)
````

````python
for this_is_an_extreme_example in name:
    print(this_is_an_extreme_example)
````

A string counts as an interable! We'll see some better examples of iterables in the next lesson!

Basically `i` gets overwritten each pass of the loop for each letter in the variable `name`. 

Everything that is indented after the `:` is called the "body" of the loop. This tells python to take that placeholder variable, `i`, and do stuff to it. In this example, we're simply printing it out.

---

`range()` function. 

To also showcase `for` loops, we'll use the built-in `range` function.

`range` simply makes a sequence of integers starting at 0 up to, but not inclusive of the number we pass through it.

````python
for i in range(10):
    print(i)
````

````python
# range only works for integers!
for i in range(10):
    print(i)
````

## While loops

`while` loops are more abstract than `for` loops when getting started.

The loop runs until some condition is not met.

psuedo-code example:

```
hungry = True

while hungry:
    if ate_food:
        hungry = False
    

````python
# often with while loops, you'll make a variable to keep track of some condition
count = 0

# while loop
while count < 10:
    print(count)
    count += 1 # += is shorthand to increment, equivalent to count = count + 1
    
````

````python
count < 10
````

What would happen if we didn't increment the `count` variable?

We would run into what's known as an *infinite loop*.

The loop will continue to run until either we manually stop the program, shutdown our computer, or the heat death of the universe.

It is **very** common to unintentionallly write infinite loops when getting started with `while` loops!
